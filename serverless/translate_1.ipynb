# AWS Translate – translate Bangla Prothom Alo articles to English

```python
# Import required libraries and set up our environment
import pandas as pd
from pprint import PrettyPrinter

import boto3
import time

print("📚 Setting up the environment...")
pp = PrettyPrinter(indent=2)
translate = boto3.client("translate")
print("✅ Environment setup complete!")
print(f"🌍 Using AWS region: {translate.meta.region_name}")
print("✅ AWS Translate client ready")
```

```python
# Load the dataset

print("📄 Loading articles.csv...")

df = pd.read_csv("articles.csv")

print("✅ Dataset loaded")
print("Columns:", list(df.columns))
print("Language counts:")
print(df["language"].value_counts())
```

```python
# Preserve original text and store translations
if "text_orig_ln" not in df.columns:
    df["text_orig_ln"] = pd.NA

if "translated_text" not in df.columns:
    df["translated_text"] = pd.NA
```

```python

print("🔁 Translating Bangla articles to English...")

for idx, row in df.iterrows():

    if row["language"] == "bn" and isinstance(row["article_text"], str):

        try:
            df.at[idx, "text_orig_ln"] = row["article_text"]

            response = translate.translate_text(
                Text=row["article_text"][:5000],
                SourceLanguageCode="bn",
                TargetLanguageCode="en"
            )

            df.at[idx, "translated_text"] = response["TranslatedText"]
            print(f"✅ Translated row {idx}")

            time.sleep(0.2)

        except Exception as e:
            print(f"❌ Translation failed at row {idx}: {str(e)}")

    elif row["language"] == "en":
        df.at[idx, "translated_text"] = row["article_text"]
```

```python
print("💾 Saving translated dataset...")

df.to_csv("articles.csv", index=False)

print("✅ Translation complete and saved")
```
